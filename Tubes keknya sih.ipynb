# 4. Hypothesis Test

```python
import math
import scipy.stats as st
import numpy as np
"""
Determined before that significant values(alpha) is 0.05
"""
one_sided_zVal_right = st.norm.ppf(.95)
one_sided_zVal_left = st.norm.ppf(.05)
two_sided_zVal = st.norm.ppf(.975)

print("One sided zVal right " + str(one_sided_zVal_right))
print("One sided zVal left " + str(one_sided_zVal_left))
print("Two sided zVal " + str(two_sided_zVal))
```

## A. Menjawab "Nilai rata-rata daerah diatas 4700 ?"

#### Menentukan Hipotesis 0 dan Hipotesis alternatif
- H0 = μ > 4700
- H1 = μ < 4700

#### Menentukan tingkat signifikan
- α = 0.05

#### Menentukan uji statistik yang digunakan dan nilai kritis
- Karena hipotesis alternatif adalah 1 sided, rata-rata dan standar deviasi dari populasi juga diketahui, jadi akan digunakan z-test dengan nilai kritikal z < -1.6449 


```python
# Proses mengambil data mean dan standar deviasi dari populasi
df_daerah = data["Daerah"]
# Inisialisasi data hipotesis
hypMean = 4700
popStd = df_daerah.std()
# Proses mengambil sampel, sebanyak 150
df_daerah_sample = np.random.choice(df_daerah, 218)
df_daerah_sample_mean = df_daerah_sample.mean()
# Mengambil z-score
z_score = (df_daerah_sample_mean - hypMean) / (popStd / math.sqrt(218))

print("z_score_sample adalah = " + str(z_score))
print("p_value_sample adalah = " + str(st.norm.sf(z_score)))
```

#### Penentuan hasil
Didapat bahwa p_value_sample 0.20203097971703676 secara signifikan jauh lebih besar dari koefisien alpha yaitu 0.05 
dan z_score_sample berada di luar daerah kritis (z < - 1.6449) yaitu 0.8343887415306006 sehingga bisa disimpulkan bahwa
### H0 dapat diterima, pernyataan nilai rata-rata daerah di atas 4700 *BENAR*

## B. Menjawab "Nilai rata-rata sumbu utama tidak sama dengan 116"

#### Menentukan Hipotesis 0 dan Hipotesis alternatif
- H0 = μ  = 116
- H1 = μ ≠ 116

#### Menentukan tingkat signifikan
- α = 0.05

#### Menentukan uji statistik yang digunakan dan nilai kritis
- Karena hipotesis alternatif adalah 2 sided, rata-rata dan standar deviasi dari populasi juga diketahui, jadi akan digunakan z-test dengan daerah kritikal z < -1.959963984540054 atau z > 1.959963984540054


```python
# Proses mengambil data mean dan standar deviasi dari populasi
df_SumbuUtama = data["SumbuUtama"]
# Inisialisasi data hipotesis
hypMean = 116
popStd = df_SumbuUtama.std()
# Proses mengambil sampel, sebanyak 150
df_SumbuUtama_sample = np.random.choice(df_SumbuUtama, 218)
# Proses mengambil rata-rata dari sample
df_SumbuUtama_sample_mean = df_SumbuUtama_sample.mean()
# Mengambil z-score
z_score = (df_SumbuUtama_sample_mean - hypMean) / (popStd / math.sqrt(218))

print("z_score_sample adalah = " + str(z_score))
print("p_value_sample adalah = " + str(st.norm.sf(z_score) * 2))
```

## C. Menjawab "Nilai Rata-rata 20 baris pertama kolom Sumbu Kecil bukan 50 ? "

#### Menentukan Hipotesis 0 dan Hipotesis alternatif
- H0 = μ  = 50
- H1 = μ ≠ 50

#### Menentukan tingkat signifikan
- α = 0.05

#### Menentukan uji statistik yang digunakan dan nilai kritis
- Karena hipotesis alternatif adalah 2 sided, rata-rata dan standar deviasi dari populasi juga diketahui. Namun, ukuran sampel lebih kecil dari 30 (dianggap tidak berdistribusi Normal) dan data tidak dipilih secara random. 
- Nilai kritis untuk t-test 2 sided dengan tingkat signifikan 0.05 adalah t < -2.093024 atau t > 2.093024


```python
# Proses mengambil data mean dari 20 data populasi pertama
df_SumbuKecil = data["SumbuKecil"][:20]
# Inisialisasi data hipotesis
hypMean = 50
# Proses mengambil 218 data sampel, sebanyak 20 data pertama
df_SumbuKecil_sample = np.random.choice(df_SumbuKecil, 218)[:20]
# Proses mengambil standard deviation dan rata-rata dari sample
sampleStd = df_SumbuKecil_sample.std()
df_SumbuKecil_sample_mean = df_SumbuKecil_sample.mean()
# Mengambil t-score 
t_score = (df_SumbuKecil_sample_mean - hypMean) / (sampleStd / math.sqrt(20))
# Mengambil p-score
p_val_from_tscore = st.t.sf(abs(t_score), 19) 
print("t_score_sample adalah = " + str(t_score))
print("p_value_sample adalah = {0:.7f}" .format(p_val_from_tscore * 2))
```

#### Penentuan hasil
Didapat bahwa p_value_sample 0.0000429 jauh lebih kecil dari koefisien alpha yaitu 0.05 
dan t_score_sample berada di dalam daerah kritis (t < -2.093024 atau t > 2.093024) yaitu 5.278134896642716 disimpulkan bahwa

### H0 ditolak , pernyataan "Nilai Rata-rata 20 baris pertama kolom Sumbu Kecil bukan 50 ? " BENAR

## D. Menjawab "Proporsi nilai Diameter yang lebih dari 85, adalah tidak sama dengan 15% "

#### Menentukan Hipotesis 0 dan Hipotesis alternatif
- H0 = p  = 15 %
- H1 = p ≠ 15 %

#### Menentukan tingkat signifikan
- α = 0.05

#### Menentukan uji statistik yang digunakan dan nilai kritis
- Karena hipotesis alternatif adalah 2 sided, rata-rata dan standar deviasi dari populasi juga diketahui, jadi akan digunakan proportion test dengan metode z-socre dengan daerah kritikal z < -1.959963984540054 atau z > 1.959963984540054


```python
# Inisialisasi p0 dan q0 
n_population_decimal_hypotesis = 0.15 #p0
complement_npopdec_hypothesis = 1 - n_population_decimal_hypotesis #q0
# Proses mengambil sampel
df_sample = data.sample(n = 218)
# Mengambil jumlah nilai diameter yang lebih dari 85 di kolom Diameter dari sampel
df_sample_processed = df_sample.loc[df_sample.Diameter > 85]
n_sample = len(df_sample_processed)
# Mencari p-complement
n_sample_decimal = n_sample / len(df_sample)
z_score = (n_sample_decimal - n_population_decimal_hypotesis) / (math.sqrt((n_population_decimal_hypotesis * complement_npopdec_hypothesis) / len(df_sample)))
print("z_score_sample adalah = " + str(z_score))
print("p_value_sample adalah = " + str(st.norm.sf(z_score) * 2))
```

#### Penentuan hasil
Didapat bahwa p_value_sample 2.712393520723168 jauh lebih kecil dari koefisien alpha yaitu 0.05 
dan z_score_sample berada di dalam daerah kritis (z < -1.959963984540054 atau z > 1.959963984540054) yaitu 2.712393520723168 disimpulkan bahwa

### H0 ditolak, pernyataan "Proporsi nilai Diameter yang lebih dari 85, adalah tidak sama dengan 15% " *BENAR  (*karena p ≠ 15 % )
