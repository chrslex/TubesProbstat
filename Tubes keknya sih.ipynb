```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# IMPORT FILE NI BOS
data = pd.read_csv("gandum.csv", header = None)
data.set_axis(["Id", "Daerah", "SumbuUtama", "SumbuKecil", "Keunikan", "AreaBulatan", "Diameter", "KadarAir"
              , "Keliling", "Bulatan", "Ransum", "Kelas"], axis = 1, inplace = True)
data.set_index("Id", inplace = True)
data.head()
```

```python
# 1. Descriptive Analysis untuk setiap kolom
print("MEAN untuk setiap kolom") 
print()
print(data.mean())
print()
print("MEDIAN untuk setiap kolom")
print()
print(data.median())
print()
print("MODUS untuk setiap kolom")
data.mode().iloc[0]
print()
print("Standar deviasi untuk setiap kolom")
print()
print(data.std())
print()
print("Variansi untuk setiap kolom")
print()
print(data.var())
print()
print("Range untuk setiap kolom")
print()
print(data.max()  - data.min())
print()
print("Nilai minimum untuk setiap kolom")
print(data.min())
print()
print("Nilai maksimum untuk setiap kolom")
print()
print(data.max())
print()
print("Kuartil untuk setiap kolom")
data.describe().iloc[4:7]
print()
print("IQR untuk setiap kolom")
Q3 = data.quantile(0.75)
Q1 = data.quantile(0.25)
print(Q3 - Q1)
print()
print("Skewness untuk setiap kolom")
print()
print(data.skew())
print()
print("Kurtosis untuk setiap kolom")
print()
print(data.kurt())
```
